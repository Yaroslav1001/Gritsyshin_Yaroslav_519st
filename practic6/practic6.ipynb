```python
import matplotlib.pyplot as plt 
import numpy as np
%matplotlib inline
N = 128
t = np.linspace(0, 1, N, endpoint=True)
x = np.sin(2*np.pi*t)
x_rect = np.zeros(N) # прямоугольный сигнал
x_rect[64:96] = 1
```

```python
plt.figure(figsize = (16, 8))
plt.subplot(1, 2, 1)
plt.plot(t, x,'r')
plt.ylabel('Amplitude')
plt.subplot(1, 2, 2)
plt.plot(x_rect)
plt.show
```

```python
def DFT(x):
    N = len(x)
    n = np.arange(N)
    k = n.reshape((N, 1))
    e = np.exp(-2j * np.pi * k * n / N)
    X = np.dot(e, x)
    return X
```

```python
def invers_DFT(x):
    N = len(x)
    n = np.arange(N)
    k = n.reshape((N, 1))
    e = np.exp(-2j * np.pi * k * n / N)
    X = np.dot(e, x)
    return X * 1/N
```
